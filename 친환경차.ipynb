```python
def make_wordcloud(text=None) : 

    from konlpy.tag import Komoran
    from collections import Counter
    import matplotlib.pyplot as plt
    from wordcloud import WordCloud


    stop_words = ['생각', '적절', '필요', '포함', '아니', '못하', '현재', '기준', '경우', '생애주', '친환경적이지', '해당', '대하', '단시간내', '적절치', '사료됨', '이후', '아무렇', '우리나라', '100', '있기때문', '이러하', '움직이', '위하', '빠르', '다가오', '정도', '늦추', '그렇','생각함','때문','동안', '충전속', '외산', '전환전', '상황', '저변', '하이브리드차', '하이브리드차량','부족한','1.','2.','3.']    
    nouns_adj_list= []
    hannanum = Komoran()
    for sentence in text:
        morph = hannanum.nouns(sentence)        
        for sentence1 in morph:
            if len(sentence1)>1:
                if sentence1 not in stop_words:
                    nouns_adj_list.append(sentence1)    

                        
                        
    count = Counter(nouns_adj_list)
    tags = count.most_common(50)
    
    wordcloud = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',  contour_width=1,contour_color='white',
                          background_color='black',colormap='Accent_r', width=1920,height=1020).generate_from_frequencies(dict(tags))

    fig = plt.figure(figsize=(12,12))
    plt.axis('off')
    plt.imshow(wordcloud)
    return tags
```

```python
from konlpy.tag import Hannanum
okt = Hannanum()
survey['Q2_new'] = survey['Q2'].apply(lambda x:okt.nouns(x))
```

```python
survey['Q2_new'] =survey['Q2_new'].apply(lambda x: list(set(x)))
```

```python
survey['Q2_new']
```

```python
c1_list = []
clean1_list = []
for i in survey['Q2_new']:
    c1_list = []
    for j in i:
        if len(j) > 1:
            c1_list.append(j)
    clean1_list.append(c1_list)
```

```python
from apyori import apriori
#연관분석을 지지도 0.15이상인 데이터들로만 구성
result = (list(apriori(clean1_list, min_support=0.05)))
```

```python
df = pd.DataFrame(result)
df['length'] = df['items'].apply(lambda x: len(x))
```

```python
df.loc[df['length']>1]
```

```python

df = df[(df['length']==2)].sort_values(by='support',ascending=False)
df_new = df
df_new
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
#네트워크 그래프 정의
G = nx.Graph()
ar = df_new['items']
G.add_edges_from(ar)
#노드 순위를 제공하는 페이지랭크를 통해 노드 색상과 크기를 지정
pr = nx.pagerank(G)
nsize = np.array([v for v in pr.values()])
nsize = 2000 * (nsize-min(nsize)/(max(nsize)-min(nsize)))
pos = nx.spring_layout(G)
import matplotlib.font_manager as fm

from matplotlib import rc

font_name = fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()

rc('font', family=font_name)


plt.figure(figsize=(16,12))
plt.axis('off')
nx.draw_networkx(G, font_family= font_name,font_size=16,
                pos = pos,node_size = nsize,
                alpha = 0.7 , edge_color = '.5', cmap=plt.cm.YlGn, node_color='#FF1744')
#plt.savefig("C:/Users/data/Desktop/교통/텍스트분석/el_total.jpg", dpi=1000)
```
